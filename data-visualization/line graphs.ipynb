```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\lic004\\Desktop\\GitHubTesting\\data\\cleanedData.csv')
df.head()
```

```python
df['OpenCloseAvg'] = (df['Open'] + df['Close'])/2
df.head()
```

```python
# Takes a while!
plt.figure(figsize=(12,4))
sns.lineplot(x='Date', y='OpenCloseAvg', data=df, hue='Company', palette='hls')
```
