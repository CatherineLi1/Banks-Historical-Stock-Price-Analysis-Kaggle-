```python
import pandas as pd
import numpy as np

import os
import zipfile
#from io import StringIO
```

```python
# create relevant objects to use later
df = pd.DataFrame()
df = df.append({'Date':np.nan,'High':np.nan,'Low':np.nan,'Open':np.nan,'Close':np.nan,'Volume':np.nan,'Adj Close':np.nan},
               ignore_index = True)
companies = set()
allCompanyLastDay = []
```

```python
# all data of last day is saved into a set of company names and a list of their data
with zipfile.ZipFile('archive.zip') as z:
    for filename in z.namelist():
        if not os.path.isdir(filename):
            # read the file
            with z.open(filename) as f:
                companies.add(filename[:-4])
                for line in f:
                    string = line.decode('utf-8')
                    if '2020-11-20' in string:
                        allCompanyLastDay.append(string)
```

```python
# remove \r\n at the end of each string, separate by commas
for i in range(len(allCompanyLastDay)):
    allCompanyLastDay[i] = allCompanyLastDay[i][:-2]
    allCompanyLastDay[i] = allCompanyLastDay[i].split(',')
allCompanyLastDay[0]
```

```python
df
```

```python
# add stocks data into dataframe
for i in range(len(allCompanyLastDay)):
    df = df.append(pd.Series(allCompanyLastDay[i], index=df.columns), ignore_index=True)
df.head()
```

```python
# remove first row of NaN values
df.drop(index=df.index[0], axis=0, inplace=True)
df.head()
```

```python
# convert companies into list and add into dataframe
companies = list(companies)
companies.sort()

df['Company'] = companies
df.head()
```

```python
df.to_csv('allStocksLastDay.csv')
```
