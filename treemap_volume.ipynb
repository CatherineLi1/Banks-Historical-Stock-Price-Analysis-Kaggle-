```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('C:\\Users\\lic004\\Desktop\\GitHubTesting\\data\\cleanedData.csv')
df.head(-5)
```

```python
lastDay = df[df['Date']=='11/20/2020']
lastDay
```

```python
import squarify
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
figure(figsize=(12,10))
sizes = lastDay['Volume']
labels = lastDay['Company']
colors=['royalblue','palevioletred','goldenrod','plum','mediumaquamarine','yellowgreen','yellow','orange','lightgreen']
squarify.plot(sizes = sizes, label=labels, alpha=0.6, color=colors)
plt.title('Treemap of Bank Volumes on 11/20/2020')
plt.show()
```
